```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv("/content/drive/My Drive/CodingDojo/Codigos/week12/nba.csv")
df.head()
```

```python
engine = create_engine("sqlite:///df.db", echo = True)
conn   = engine.connect()
```

```python
sqlite_table = "nba_details"
df.to_sql(sqlite_table, conn)
```

```python
sql = """
select * 
from nba_details"""

df_sql = pd.read_sql_query (sql,engine)
df_sql.head()
```

```python
sql = """
select Name, FGA, 1 as auxiliar
from nba_details"""

df_sql = pd.read_sql_query (sql,engine)
df_sql.head()
```

```python
sql = """
select Name, avg(GP) as media
from nba_details
group by name
"""

df_sql = pd.read_sql_query (sql,engine)
df_sql.head()
```

```python
sql = """
select *
, 2022 as aux
, avg(GP) as media
from nba_details
where MIN > 0 and name like '%Morrison%'
group by name
"""

df_sql = pd.read_sql_query (sql,engine)
df_sql.head()
```

```python
sql = """
select *
, length(name) as n_elementos
, substr(name,5, length(name)) as n_5
from nba_details
"""

df_sql = pd.read_sql_query (sql,engine)
df_sql.head()
```
